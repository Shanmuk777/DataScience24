```python
# 1. Reverse a string
s = "hello"
reversed_s = s[::-1]
print(reversed_s)  # Output: "olleh"

# 2. Count the number of vowels in a string
s = "hello world"
vowels = "aeiouAEIOU"
vowel_count = sum(1 for char in s if char in vowels)
print(vowel_count)  # Output: 3

# 3. Check if a given string is a palindrome
s = "madam"
is_palindrome = s == s[::-1]
print(is_palindrome)  # Output: True

# 4. Check if two given strings are anagrams
s1 = "listen"
s2 = "silent"
are_anagrams = sorted(s1) == sorted(s2)
print(are_anagrams)  # Output: True

# 5. Find all occurrences of a substring
s = "banana"
sub = "an"
occurrences = [i for i in range(len(s)) if s.startswith(sub, i)]
print(occurrences)  # Output: [1, 3]

# 6. Basic string compression
s = "aaabbc"
compressed = []
count = 1
for i in range(1, len(s)):
    if s[i] == s[i - 1]:
        count += 1
    else:
        compressed.append(s[i - 1] + str(count))
        count = 1
compressed.append(s[-1] + str(count))
compressed_s = "".join(compressed)
print(compressed_s)  # Output: "a3b2c1"

# 7. Check if a string has all unique characters
s = "abcdef"
is_unique = len(set(s)) == len(s)
print(is_unique)  # Output: True

# 8. Convert string to uppercase or lowercase
s = "Hello"
upper_s = s.upper()
lower_s = s.lower()
print(upper_s, lower_s)  # Output: "HELLO hello"

# 9. Count the number of words in a string
s = "Hello world! This is Python."
word_count = len(s.split())
print(word_count)  # Output: 5

# 10. Concatenate two strings without using +
s1 = "Hello"
s2 = "World"
concatenated_s = "{}{}".format(s1, s2)
print(concatenated_s)  # Output: "HelloWorld"
```

```python
# 11. Remove all occurrences of a specific element from a list
lst = [1, 2, 3, 2, 4]
elem = 2
filtered_lst = [x for x in lst if x != elem]
print(filtered_lst)  # Output: [1, 3, 4]

# 12. Find the second largest number in a list
lst = [10, 20, 4, 45, 99]
unique_lst = list(set(lst))
unique_lst.sort()
second_largest = unique_lst[-2] if len(unique_lst) > 1 else None
print(second_largest)  # Output: 45

# 13. Count occurrences of each element in a list
lst = [1, 2, 2, 3, 3, 3]
count_dict = {x: lst.count(x) for x in set(lst)}
print(count_dict)  # Output: {1: 1, 2: 2, 3: 3}

# 14. Reverse a list in-place
lst = [1, 2, 3, 4]
left, right = 0, len(lst) - 1
while left < right:
    lst[left], lst[right] = lst[right], lst[left]
    left += 1
    right -= 1
print(lst)  # Output: [4, 3, 2, 1]

# 15. Remove duplicates while preserving order
lst = [1, 2, 2, 3, 4, 4]
seen = set()
deduplicated_lst = [x for x in lst if not (x in seen or seen.add(x))]
print(deduplicated_lst)  # Output: [1, 2, 3, 4]

# 16. Check if a list is sorted
lst = [1, 2, 3]
is_sorted = lst == sorted(lst) or lst == sorted(lst, reverse=True)
print(is_sorted)  # Output: True

# 17. Merge two sorted lists
lst1 = [1, 3, 5]
lst2 = [2, 4, 6]
merged_list = sorted(lst1 + lst2)
print(merged_list)  # Output: [1, 2, 3, 4, 5, 6]

# 18. Find the intersection of two lists
lst1 = [1, 2, 3]
lst2 = [2, 3, 4]
intersection = list(set(lst1) & set(lst2))
print(intersection)  # Output: [2, 3]

# 19. Find the union of two lists without duplicates
union = list(set(lst1) | set(lst2))
print(union)  # Output: [1, 2, 3, 4]

# 20. Shuffle a list randomly
import random
lst = [1, 2, 3, 4, 5]
for i in range(len(lst) - 1, 0, -1):
    j = random.randint(0, i)
    lst[i], lst[j] = lst[j], lst[i]
print(lst)  # Output: Randomly shuffled list
```

```python
# 21. Find common elements between two tuples
t1 = (1, 2, 3)
t2 = (2, 3, 4)
common_elements = tuple(set(t1) & set(t2))
print(common_elements)  # Output: (2, 3)

# 22. Prompt user for two sets and find intersection
set1 = {1,2,3,4}
set2 = {3,4,5,6}
print("Intersection:", set1 & set2)

# 23. Concatenate two tuples
t1 = (1, 2)
t2 = (3, 4)
concatenated_tuple = t1 + t2
print(concatenated_tuple)  # Output: (1, 2, 3, 4)

# 24. Elements in first set but not second
set1 = {"a", "b", "c"}
set2 = {"b", "c", "d"}
difference = set1 - set2
print(difference)  # Output: {'a'}

# 25. Extract elements from a tuple within a range
tpl = (10, 20, 30, 40, 50)
start, end = 1, 4
sliced_tpl = tpl[start:end]
print(sliced_tpl)  # Output: (20, 30, 40)

# 26. Union of two sets
print(set1 | set2)  # Output: {'a', 'b', 'c', 'd'}

# 27. Max and Min of a tuple
tpl = (3, 1, 4, 1, 5, 9)
min_val, max_val = min(tpl), max(tpl)
print(min_val, max_val)  # Output: 1 9

# 28. Set operations
print(set1 | set2, set1 & set2, set1 - set2)

# 29. Count occurrences in a tuple
tpl = (1, 2, 2, 3, 4, 2)
count_2 = tpl.count(2)
print(count_2)  # Output: 3

# 30. Symmetric difference of sets
print(set1 ^ set2)  # Output: {'a', 'd'}
```
